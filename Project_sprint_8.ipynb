# Prueba de hipótesis

En esta sección se analizará si las condiciones climáticas influyen en la duración de los viajes en taxi desde el Loop hasta el Aeropuerto Internacional O'Hare. Para ello, se utilizará un conjunto de datos que contiene la fecha y hora de inicio del viaje, las condiciones climáticas y la duración del recorrido en segundos.

El objetivo es determinar, mediante una prueba estadística, si la duración promedio de los viajes realizados durante los sábados lluviosos difiere de la duración promedio de los viajes realizados en sábados sin lluvia.

```python
import pandas as pd 
df3 = pd.read_csv('/datasets/project_sql_result_07.csv')
```

## Exploración de datos

Antes de realizar la prueba de hipótesis, se llevará a cabo una exploración inicial del conjunto de datos para comprender su estructura y verificar su calidad.

Para ello, se utilizarán los métodos .info(), .head() y .describe(). Con .info() se revisará el número de registros, los tipos de datos y la existencia de valores nulos. Mediante .head() se visualizarán las primeras filas del dataset para conocer la organización de las variables, mientras que .describe() permitirá obtener estadísticas descriptivas de las variables numéricas.

Esta revisión es importante para confirmar que los datos son consistentes y adecuados para realizar el análisis estadístico y la prueba de hipótesis.

```python
print(df3.head())
print(df3.info())
print(df3.describe())
```

```python
df3['start_ts'] = pd.to_datetime(df3['start_ts'])
print(df3)
```

```python
df_saturday = df3[df3['start_ts'].dt.dayofweek == 5]
print(df_saturday)
```

Una vez revisado el dato de 1068 registros, que coincide exactamente con el tamaño del dataset original, se observa que el conjunto de datos ya estaba previamente filtrado por los días sábados. Esto indica que la transformación principal de los datos se realizó en la consulta SQL original, por lo que el filtrado adicional en Python no era necesario.

Sin embargo, esto no afecta la validez de la prueba de hipótesis, ya que el análisis comparó correctamente los grupos de viajes en distintas condiciones climáticas (sábados lluviosos y no lluviosos) utilizando la variable de interés, la duración de los viajes.

Este hallazgo resalta la importancia de comprender el origen y el preprocesamiento de los datos antes de realizar análisis estadísticos, ya que permite evitar pasos redundantes y mejora la interpretación del flujo completo del análisis.

# Planteamiento de hipótesis

Para evaluar si las condiciones climáticas influyen en la duración de los viajes, se plantean las siguientes hipótesis:

- **Hipótesis nula (H₀):** La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare es igual en sábados lluviosos y en sábados sin lluvia.

- **Hipótesis alternativa (H₁):** La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare es diferente en sábados lluviosos y en sábados sin lluvia.

**Se establece un nivel de significancia de α = 0.05.**

Este valor indica que se acepta un 5% de probabilidad de cometer un error al rechazar la hipótesis nula, es decir, concluir que existe una diferencia cuando en realidad no la hay.

```python
df_saturday['weather_conditions'].unique()
```

```python
bad_weather = df_saturday[df_saturday['weather_conditions'] == 'Bad']
good_weather = df_saturday[df_saturday['weather_conditions'] == 'Good']
print(bad_weather)
print(good_weather)
```

Para realizar la prueba de hipótesis se utilizó la prueba t para muestras independientes (ttest_ind), ya que se busca comparar la media de la duración de los viajes entre dos grupos independientes: sábados con lluvia y sábados sin lluvia.

Esta prueba es adecuada porque las observaciones de ambos grupos no están relacionadas entre sí y la variable analizada es cuantitativa continua (duración del viaje). Además, el objetivo del análisis es evaluar si existe una diferencia significativa entre las medias de ambos grupos.

El criterio de decisión se basa en el valor p: si el p-value es menor que α = 0.05, se rechaza la hipótesis nula; de lo contrario, no se rechaza.

```python
from scipy import stats
stats.ttest_ind(bad_weather['duration_seconds'], good_weather['duration_seconds'])
```

El p-value obtenido es extremadamente pequeño (6.5e-12), mucho menor que el nivel de significancia (0.05). Por lo tanto, se rechaza la hipótesis nula. Esto indica que existe evidencia estadísticamente significativa de que la duración promedio de los viajes difiere entre sábados lluviosos y no lluviosos.

El estadístico t positivo indica que la duración promedio de los viajes es mayor en los sábados con condiciones climáticas adversas (lluvia) en comparación con los sábados sin lluvia. Esto sugiere que el clima podría influir en el aumento del tiempo de los viajes, posiblemente debido a factores como menor velocidad de circulación o mayor congestión vehicular.
